```python
from Paths import Point
from Paths import PathUtilities
from Paths import MultiPathHandler
from Utilities import DistanceType
import matplotlib.pyplot as plt
#Logits
from AdaptivePathSizeLogit import AdaptivePathSizeLogitCalculator
import numpy as np
import matplotlib.pyplot as plt
```

```python
pathUtils = PathUtilities()
distanceType = DistanceType.EUCLIDEAN
#We will try our methods with 5 paths
points1 = [Point(0,-4), Point(1,-4), Point(1,-3), Point(5,-3), Point(5,-2),Point(6,-2), Point(6,1.5)]
path1 = pathUtils.getPathFromPointsAndDistance(points1,distanceType)
points2 = [Point(0,-4), Point(1,-4),Point(1,-1), Point(5,-1), Point(5,-2),Point(6,-2), Point(6,1.5)]
path2 = pathUtils.getPathFromPointsAndDistance(points2, distanceType)
points3 = [Point(0,-4), Point(-1,-4), Point(-1,1), Point(1,1),Point(1,2),Point(3.5, 2), Point(3.5,1),Point(6,1), Point(6,1.5)]
path3 = pathUtils.getPathFromPointsAndDistance(points3, distanceType)
points4 = [Point(0,-4),Point(0,-1),Point(5,-1),Point(5,0),Point(6,0),Point(6,1.5)]
path4 = pathUtils.getPathFromPointsAndDistance(points4, distanceType)
points5 = [Point(0,-4), Point(0,0),Point(3.5,0),Point(3.5,1),Point(5,1),Point(5,0),Point(6,0),Point(6,1.5)]
path5 = pathUtils.getPathFromPointsAndDistance(points5, distanceType)
paths = [path2,path5, path1,path3, path4]
pathHandler = MultiPathHandler(paths)
pathHandler.drawPaths()
```

```python
adaptiveLogitCalculator = AdaptivePathSizeLogitCalculator(pathHandler)
```

```python
probs = adaptiveLogitCalculator.getProbabilities()
```

```python
adaptiveLogitCalculator.showProbabilityDistribution()
```
